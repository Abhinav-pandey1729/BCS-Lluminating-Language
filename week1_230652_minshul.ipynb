```python
#drop duplicate values
dataset.duplicated().sum()
dataset = dataset.drop_duplicates()
dataset
```

```python
#changing case to lowercase
dataset['text']=dataset['text'].str.lower()
dataset
```

```python
#stopword removal
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
stopwords.words('english')

def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    ntext = [word for word in word_tokens if word.lower() not in stop_words]
    return ' '.join(ntext)
dataset['text'] = dataset['text'].apply(remove_stopwords)
```

```python
dataset.head()
```

```python
#stemming the text
from nltk.stem.porter import PorterStemmer

def stemming(text):
    stemmer = PorterStemmer()
    word_tokens = word_tokenize(text)
    ntext= [stemmer.stem(word) for word in word_tokens]
    return ' '.join(ntext)
```

```python
dataset['text'] = dataset['text'].apply(stemming)
dataset.head()
```

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

def lemmatize_text(text):
    lemmatizer = WordNetLemmatizer()
    tokens = word_tokenize(text)
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return ' '.join(lemmatized_tokens)
```

```python
dataset['text'] = dataset['text'].apply(lemmatize_text)
dataset.tail()
```

```python
#test train split and then tokenization
from sklearn.model_selection import train_test_split
X=dataset['text']
Y=dataset['label']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.30,random_state=0)
```

```python
#trained=X_train
tokenizer=Tokenizer(num_words=3000000)
tokenizer.fit_on_texts(X_train)
word_index=tokenizer.word_index
print(word_index)
```

```python
#feature extraction
from sklearn.feature_extraction.text import TfidfVectorizer
vect = TfidfVectorizer().fit(X_train)
X_train_vectorized = vect.transform(X_train)
X_test_vectorized = vect.transform(X_test)
```
